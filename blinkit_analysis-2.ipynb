```python
df['Item Fat Content'] = df['Item Fat Content']. replace({'LF': 'Low Fat' ,
                                                          'low fat': 'Low Fat',
                                                          'reg': 'Regular'
                                                         })
```

```python
#**BUSINESS REQUIREMENT**
```

```python
#Total Sales
total_sales = df['Sales'].sum()

#Average Sales
avg_sales = df[ 'Sales']. mean()

#No of Items Sold
no_of_items_sold = df['Sales'].count()

#Average Ratings
avg_ratings = df[ 'Rating'].mean ()


#Display

print(f"Total Sales: $(total_sales:,.0f)")
print(f"Average Sales: $(avg_sales:,.0f)")
print(f"No of Items Sold: {no_of_items_sold:,.0f}")
print(f"Average Ratings: {avg_ratings:,.1f}")
```

```python
# CHARTS REQUIREMENTS 
```

```python
## TOTAL SALE BY FAT CONTENT
```

```python
# Group sales by fat content
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

# Plot a pie chart
plt.pie(
    sales_by_fat, 
    labels=sales_by_fat.index,
    autopct='%.1f%%',   # Fixed syntax
    startangle=90
)

plt.title('Sales by Fat Content')
plt.axis('equal')  # Equal aspect ratio ensures pie is a circle
plt.show()
```

```python
## TOTAL SALE BY ITEM TYPE 
```

```python
# Group sales by item type and sort descending
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

# Create figure
plt.figure(figsize=(10, 6))

# Create bar chart
bars = plt.bar(sales_by_type.index, sales_by_type.values)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=90)  # Use 90 instead of -90 for clarity

# Add labels and title
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

# Add value labels on top of bars
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # X coordinate (center of the bar)
        bar.get_height(),                     # Y coordinate (top of the bar)
        f'{bar.get_height():,.0f}',           # Text with commas and no decimals
        ha='center', va='bottom', fontsize=8
    )

plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python
## fat content by outlet for total sales 
```

```python
# Group sales by Outlet Location Type and Item Fat Content
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()

# Ensure columns are in the desired order
grouped = grouped[['Regular', 'Low Fat']]

# Plot a bar chart
ax = grouped.plot(
    kind='bar',
    figsize=(8, 5),
    title='Outlet Tier by Item Fat Content'
)

# Labels and legend
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
##TOTAL SALES BY OUTLET ESTABLISHMENT
```

```python
# Group sales by Outlet Establishment Year
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

# Plot line chart
plt.figure(figsize=(9,5))
plt.plot(sales_by_year.index, sales_by_year.values,
         marker='o', linestyle='-')  # Use 'o' for marker

# Labels and title
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment Year')

# Annotate values on the line
for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```
